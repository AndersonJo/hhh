```python
import numpy as np
import torch
```

$$ f(x) = x^2 $$

```python
x_np = 3.0
y_np = x_np ** 2
grad_np = 2 * x_np

print('x :', x_np)
print('y :', y_np)
print('grad :', grad_np)

# Pytorch 에서 자동으로 미분을 계산함
x_pt = torch.tensor(3.0, requires_grad=True)
y_pt = x_pt ** 2
y_pt.backward()

print('x :', x_pt.item())
print('y :', y_pt.item())
print('grad :', x_pt.grad.item())
```
